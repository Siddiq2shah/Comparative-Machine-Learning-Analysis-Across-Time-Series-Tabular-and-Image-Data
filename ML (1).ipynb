```python
import pandas as pd

filename = "adult.data.txt"

columns = [
    "age","workclass","fnlwgt","education","education_num",
    "marital_status","occupation","relationship","race","sex",
    "capital_gain","capital_loss","hours_per_week","native_country","income"
]

df = pd.read_csv(filename, header=None, names=columns, sep=",", skipinitialspace=True)

df.head()
```

```python
# Clean income labels (remove periods)
df["income"] = df["income"].str.replace(".", "", regex=False)

# Replace '?' with proper missing values
df = df.replace("?", pd.NA)

# Save spreadsheet versions
df.to_csv("adult_clean.csv", index=False)
df.to_excel("adult_clean.xlsx", index=False)

df.shape   # optional: shows dimensions of the dataset
```

```python
import bz2

# Decompress training file
with bz2.open("mnist.scale.bz2", "rb") as f_in:
    with open("mnist.scale", "wb") as f_out:
        f_out.write(f_in.read())

# Decompress test file
with bz2.open("mnist.scale.t.bz2", "rb") as f_in:
    with open("mnist.scale.t", "wb") as f_out:
        f_out.write(f_in.read())
```

```python
from sklearn.datasets import load_svmlight_file
import pandas as pd

X_train, y_train = load_svmlight_file("mnist.scale")

X = X_train.toarray()
df_mnist = pd.DataFrame(X)
df_mnist.insert(0, "label", y_train.astype(int))

df_mnist.head()
```

```python
df_mnist.to_csv("mnist_clean.csv", index=False)
df_mnist.to_excel("mnist_clean.xlsx", index=False)
```
